```python
import whisperx
import gc
```

```python
device = "cuda"
batch_size = 16
compute_type = "float16"
```

```python
!pwd
```

```python
# https://www.youtube.com/watch?v=0jNtb-nchZU&ab_channel=%E3%81%8D%E3%82%88
audio_file = "aya.wav"
```

```python
model = whisperx.load_model("large-v2", device, compute_type=compute_type)
```

```python
audio = whisperx.load_audio(audio_file)
result = model.transcribe(audio, batch_size=batch_size)
print(result['segments'])
```

```python
result
```

## Word Level Transcribe

```python
model_a, metadata = whisperx.load_align_model(language_code=result["language"], device=device)

result_by_words = whisperx.align(result['segments'], model_a,
                        metadata,
                        audio,
                        device,
                        return_char_alignments=False)

print(result_by_words['segments'])
```

## Identify Multi Speakers

```python
device = "cuda"
batch_size = 16
compute_type = "float16"
```

```python
diarize_model = whisperx.DiarizationPipeline(device=device)
```

```python
diarize_segments = diarize_model(audio, min_speakers=4)
```

```python
diarize_segments
```

```python
diarize_segments.speaker.unique()
```

```python
res = whisperx.assign_word_speakers(diarize_segments, result)
print(diarize_segments)
# print(res['segments'])
```
